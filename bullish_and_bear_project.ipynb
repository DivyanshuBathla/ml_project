```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
path ='C:\\Users\\as\\Desktop\\csv_files'
csv_files = glob.glob(os.path.join(path, "*.csv"))

```

```python
data=[]
for f in csv_files:
    data.append(pd.read_csv(f))
        
list=('BharatPetroleum_','HindustanUniliver_','Reliance_','TataSteel_')
```

```python
ran=len(data)
ran=4
list[0]
```

```python

df=data[0]
st=df[['High','Low','Close']].values
st
stock=[st]
for i in range(1,ran):
    df=data[i]
    st=df[['High','Low','Close']].values
    stock.append(st)

```

```python
high_data=[]
low_data=[]
for i in stock:
    high_data.append(i[:,0:1])
    low_data.append(i[:,1:2])
```

```python
def observation(day,arr,num,count,strin,count_trend):
    if num>-8:
        trend=0
        inc=0
        dec=0
        per_inc=0.0
        per_dec=0.0
        inc_count=0
        dec_count=0
        i=day+1
        j=day
        start=0    
        while(trend<2 and day<count and i<count):
           if(arr[i]>arr[j]):
           
           
              if(trend==0):
                start=1
              trend+=1
              while(arr[i]-arr[j]>-3):
                 inc+=1
                 if arr[i]>arr[j]:
                    inc_count+=1
                    j=i
                    i+=1
                    if(i>count):
                       break
                 else:
                     i+=1
                     if(i>count):
                         break
           else:
           
           
             if(trend==0):
              start=-1
             trend+=1
             while(arr[i]-arr[j]<3):
               dec+=1
               if(arr[i]<arr[j]):
                   dec_count+=1
                   j=i
                   i+=1
                   if(i>count):
                       break
               else:
                    i+=1
                    if(i>count):
                       break
        file=open(list[num]+strin,'a')       
        st=stock[num]
        if(start==1):
          if inc_count>8:
           file.write("\nDrastic Increase for ")
           file.write(str(inc_count))
           file.write(" days")
          elif inc_count>5:
            file.write("\nImediate Increase for ")
            file.write(str(inc_count))
            file.write(" days")
          else:
            file.write("\nSlight Increase for ")
            file.write(str(inc_count))
            file.write(" days")
          file.write("\nPrice after increase=Rs ")
          file.write(str(round(st[day+inc,2],2)))
          per_inc=(arr[day+inc]-arr[day])/arr[day]*100
          file.write("\nPercantage increase=")
          x=per_inc[0]
          x=round(x,2)
          file.write(str(x))
          file.write('%')
          if dec_count>8:
           file.write("\nthen Drastic Decrease for ")
           file.write(str(dec_count))
           file.write(" days")
          elif dec_count>5:
            file.write("\nthen Imediate Decrease for ")
            file.write(str(dec_count))
            file.write(" days")
          else:
            file.write("\nthen Slight Decrease for ")
            file.write(str(dec_count))
            file.write(" days")
          file.write("\nPrice after decrease=Rs ")
          file.write(str(round(st[day+inc+dec,2],2)))
          per_dec=(arr[day+inc]-arr[day+inc+dec])/arr[day+inc]*100
          file.write("\nPercantage decrease=")
          x=per_dec[0]
          x=round(x,2)
          file.write(str(x))
          file.write('%')
          count_trend[0]=count_trend[0]+inc_count
          count_trend[1]=count_trend[1]+dec_count
        else:
          if dec_count>8:
            file.write("\nDrastic Decrease for ")
            file.write(str(dec_count))
            file.write(" days")
          elif dec_count>5:
            file.write("\nImediate Decrease for ")
            file.write(str(dec_count))
            file.write(" days")
          else:
            file.write("\nSlight Decrease for ")
            file.write(str(dec_count))
            file.write(" days")
          file.write("\nPrice after decrease=Rs ")
          file.write(str(round(st[day+dec,2],2)))
          per_dec=(arr[day]-arr[day+dec])/arr[day]*100
          file.write("\nPercantage decrease=")
          x=per_dec[0]
          x=round(x,2)
          file.write(str(x))
          file.write('%')
          if inc_count>8:
            file.write("\nthen Drastic Increase for ")
            file.write(str(inc_count))
            file.write(" days")
          elif inc_count>5:
            file.write("\nthen Imediate Increase for ")
            file.write(str(inc_count))
            file.write(" days")
          else:
            file.write("\nthen Slight Increase for ")
            file.write(str(inc_count))
            file.write(" days")
          file.write("\nPrice after Increase=Rs ")
          file.write(str(round(st[day+inc+dec,2],2)))
          per_inc=(arr[day+inc+dec]-arr[day+dec])/arr[day+dec]*100
          file.write("\nPercantage increase=")
          x=per_inc[0]
          x=round(x,2)
          file.write(str(x))
          file.write('%')
          count_trend[2]=count_trend[2]+dec_count
          count_trend[3]=count_trend[3]+inc_count
        file.write("\n")
        file.close()    
    return count_trend    
```
